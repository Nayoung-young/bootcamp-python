## 2-5-1. List Comprehension
- 리스트를 빠르고 쉽게 생성하는 기능
- 일반적으로 for 문을 사용하는 것보다 무척 빠름
- 프로그램 동작 시간이 너무 오래 걸리는 경우, 대안으로 생ㄱ가해볼 수 있음

- 기본 문법
  - 리스트형 = [변수 for 변수 in [변수1, 변수2, 변수3, ...]]
  - 리스트형 = [변수 for 변수 in [변수1, 변수2, 변수3, ...] if 조건문]

```python
new_list = [elem for elem in range(10)]
new_list
```

```python
odd_list =  [elem for elem in range(10) if elem%2 == 1] # 홀수만 추가
odd_list
```

```python
# for문과 성능 차이
from timeit import timeit
iteration = 10_0000

def for_loop():
  new_list = []
  for elem in range(iteration):
    if (elem % 2 ) == 1:
      new_list.append(elem)

def list_comprehension():
  new_list =  [elem for elem in range(iteration) if elem%2 == 1]
```

```python
timeit(for_loop, number = 100)
```

```python
timeit(list_comprehension, number = 100)
```

## 2.5.2 예외처리
- 예외 처리 = 프로그램 동작에 오류가 발생할 때
- 오류를 무시하거나, 오류 유형에 따라 이후 조치를 추가해주는 기능
- try와 except 문을 사용해서 나타냄
- 기본 문법
  - try : 우선적으로 실행할 구문 ...
  - except : 오류 발생 시 실행할 구문 ...

```python
def error_function(input):
  for i in range(10):
    res = input / i
    print(res)
```

```python
error_function(10)
```

> 오류 무시하기
> - pass를 사용해 모든 오류를 무시합니다
> - 치명적인 에러까지 무시하기 때문에 추천 X

```python
def error_function(input):
  for i in range(10):
    try:
      res = input / i
      print(res)
    except:
      pass
```

```python
error_function(10)
```

> 바람직한 방향
> - 예상되는 예외만을 지정해서 처리
> - except: pass로 모든 예외를 무시해버리면, 이후 문제 발생 시 원인 파악 힘듦

```python
def error_function(input):
  for i in range(10):
    try:
      res = input / i
      print(res)
    except ZeroDivisionError:
      pass
```

```python
error_function(10)
```
