```python
# 도화지를 깔고
plt.figure()

# 해당 리스트를 기준으로 하는 선을 그립니다
plt.plot([1, 2, 3, 4, 5], [1, 2, 3, 8, 16]) # 기본은 선 그래프
# plt.plot(x축, y축)

# 보여주기
plt.show()
```

## Bar plot

```python
# 막대 그래프

plt.figure(figsize = (8,6))
# plt.grid(alpha = 0.3, color = 'red') # 격자 배경 넣기

plt.title("Coffe Sales in March", fontsize = 20, loc = 'center')
plt.xlabel("Menu", fontsize = 8)
plt.ylabel("Sales", fontsize = 8)

plt.plot([100, 200, 40, 800, 1600], label = "Google Trend")
plt.bar(x, y, color = 'lightcoral', label = "Coffee Sales") # 비교
# plt.plot(x,y) # 경향

plt.legend(loc = "best") # 라벨 출력하기
# plt.ylim(0, 2000)
plt.yticks([num for num in range(0, 2000, 200)]) # y축

plt.show()
```

## Scatter plot

```python
import numpy as np

x = np.random.random(1000)
y = np.random.random(1000)

plt.figure(figsize = (6, 6))
plt.scatter(x, y, color='orange', marker = "x", s=100, alpha =0.3)
plt.show()
```

## Subplot

```python
# subplot()을 이용하여 여러 개의 plot을 하나의 figure에 출력할 수 있다.

# plt.subpot(nrows, ncols, index)
# nrows와 ncols가 있다고 했을 때의 위치!

plt.figure()

# plt.subplot(2, 1, 1)
# # plt.subplot(2, 1, 2)
# plt.subplot(2, 2, 3)
# plt.subplot(2, 2, 4)

plt.subplot(1, 2, 1)
plt.plot([0.1, 0.3, 0.5])

# plt.subplot(1, 2, 2)
plt.subplot(3, 2, 2)
plt.bar([1, 2, 5], [1, 2, 5])

plt.subplot(3, 2, 4)
plt.pie([1, 2, 3])

plt.subplot(3, 2, 6)
plt.scatter(x, y)

plt.show()
```

```python
# DataFrame으로 바로 plotting 하기

import pandas as pd

# sample data
x = ["Americano", "Cafe Latte", "Vanilla Latte", "Cold Brew", "Flat White"]
y = [1463, 301, 866, 905, 274]

df = pd.DataFrame(data = y, index = x, columns = ["Count"])

df.sort_values(by='Count', ascending=False).plot(kind = "bar")
```
